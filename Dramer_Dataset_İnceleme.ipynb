```python
# Anahtarları listele
print("Keys in the loaded data:", data.keys())

# DREAMER anahtarını detaylı inceleyelim
dreamer_data = data['DREAMER']
print("Type of DREAMER data:", type(dreamer_data))
print("Shape of DREAMER data:", dreamer_data.shape)
```

```python
# DREAMER içindeki yapılandırılmış veri
dreamer_struct = data['DREAMER'][0, 0]  # (1, 1) boyutundaki dizinin tek öğesi

# Yapılandırılmış verinin alanlarını listele
print("Fields in DREAMER structure:", dreamer_struct.dtype.names)
```

```python
# Data alanını incele
data_field = dreamer_struct['Data']
print("Type of Data field:", type(data_field))
print("Shape of Data field:", data_field.shape)
```

```python
# İlk katılımcının verisini incele
participant_1_data = data_field[0, 0]

# İlk katılımcının veri alanlarını listele
print("Fields in participant data:", participant_1_data.dtype.names)
```

```python
# Yaş ve cinsiyet bilgilerini yazdır
age = participant_1_data['Age']
gender = participant_1_data['Gender']
print("Age:", age)
print("Gender:", gender)
```

```python
# EEG verisini kontrol et
eeg_data = participant_1_data['EEG']
print("Type of EEG data:", type(eeg_data))
print("Shape of EEG data:", eeg_data.shape)
```

```python
# EEG verisinin içeriğini kontrol et
eeg_content = eeg_data[0, 0]
print("Type of EEG content:", type(eeg_content))
print("Shape of EEG content:", eeg_content.shape)
```

```python
# EEG içeriğinin daha derinini kontrol et
eeg_inner_content = eeg_content[0, 0]
print("Type of deeper EEG content:", type(eeg_inner_content))
print("Shape of deeper EEG content:", eeg_inner_content.shape)
```

```python
# numpy.void içindeki alanları listele
print("Fields in deeper EEG content:", eeg_inner_content.dtype.names)
```

```python
# Baseline verisini kontrol et
baseline_data = eeg_inner_content['baseline']
print("Type of baseline data:", type(baseline_data))
print("Shape of baseline data:", baseline_data.shape)
```

```python
# İlk kanalı incele
first_channel = baseline_data[0, 0]
print("Type of first channel data:", type(first_channel))
print("Shape of first channel data:", first_channel.shape)
```

```python
# İlk kanalın bir kısmını görüntüle
print("First 5 rows of the first channel data:")
print(first_channel[:5])  # İlk 5 satırı göster
```

```python
import matplotlib.pyplot as plt

# İlk sütunun zaman serisini çizdir
plt.plot(first_channel[:, 0])  # İlk sütun
plt.title("Time Series of the First Channel (First Feature)")
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.show()
```
